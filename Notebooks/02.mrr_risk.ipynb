```python
import pandas as pd
import matplotlib.pyplot as plt, seaborn as sns

path = '/Users/luanabreno/Desktop/Telco.2.0 - Project/'

df_clean = pd.read_parquet(path + '/data_processed/churn_clean.parquet')
```

### We closed the last notebook with this info

- High Risk customers have an average MRR of $79.59 
- There are 620 *High-Risk* active clients in the dataset 
- There are 1933 *Medium-Risk* active clients in the dataset 

```python
MRR = df_clean[df_clean['Churn'] == 'No']['MonthlyCharges'].sum()
print(f'MRR: ${MRR:,.2f}')

MRR_lost = df_clean[df_clean['Churn'] == 'Yes']['MonthlyCharges'].sum()
print(f'MRR lost due to churn: ${MRR_lost:,.2f}')

MRR_at_risk = df_clean[(df_clean['Churn'] == 'No') & 
        (df_clean['risk_segment'].isin(['High Risk', 'Medium Risk']))]['MonthlyCharges'].sum()
print(f'MRR at risk: ${MRR_at_risk:,.2f}')

MRR_saved_30 = MRR_at_risk * 0.3
print(f'MRR saved with 30% retention: ${MRR_saved_30:,.2f}')

MRR_saved_50 = MRR_at_risk * 0.5
print(f'MRR saved with 50% retention: ${MRR_saved_50:,.2f}')
```
